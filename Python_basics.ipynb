```python
df['A'] = pd.to_datetime(df['start'])
df['B'] = pd.to_datetime(df['end'])
#df['time']=df['B'] - df['A']
df['time']=(df['A'] - df['B']) / np.timedelta64(-1, 'D')
df['time']=df['time']*86400
df['count']=0
df
```

```python
df.loc[(df['A'].dt.day==d) & (df['A'].dt.month==m)& (df['A'].dt.year==yr), 'day'] = 1
day_df= df.loc[df['day']==1]
day_df
```

```python
arr = np.zeros((25,), dtype='float64')
```

```python
for index, row in day_df.iterrows():
    arr[row['id']]+=row['time']
```

```python
def humanize_time(secs):
    mins, secs = divmod(secs, 60)
    hours, mins = divmod(mins, 60)
    return '%02d hours %02d minutes %02d seconds' % (hours, mins, secs)
```

```python
index = np.zeros((24,1),dtype='int64')
j=0
for i in range(1,24):
    print (humanize_time(arr[i].astype('int')))
    if(arr[i]!=0):
        index[j]=i
        j+=1
```
